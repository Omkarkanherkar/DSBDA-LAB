```python
import pandas as pd 
import numpy as np

df=pd.read_csv('/home/ubuntu/Downloads/iris Dataset/Iris.csv')
print(df)
```

```python
mean = df['SepalWidthCm'].mean()
median = df['SepalWidthCm'].median()
mode = df['SepalWidthCm'].mode()

print("Mean of SepalWidthCm:", mean)
print("Median of SepalWidthCm:", median)
print("Mode of SepalWidthCm:", mode)
```

```python
std_A = df['SepalWidthCm'].std()
print("Standard Deviation of SepalWidthCm:", std_A)
max_A = df['SepalWidthCm'].max()
print("Maximum of SepalWidthCm:", max_A)
# Calculate minimum of column 'A'
min_A = df['SepalWidthCm'].min()
print("Minimum of SepalWidthCm:", min_A)

```

```python
import pandas as pd



# Group by 'species'
grouped = df.groupby('Species')


print("Mean for each species:")
print(grouped.mean())  
print("\n")

print("Median for each species:")
print(grouped.median())  
print("\n")


print("Standard Deviation for each species:")
print(grouped.std())  
print("\n")


print("Minimum for each species:")
print(grouped.min()) 
print("\n")


print("Maximum for each species:")
print(grouped.max()) 
print("\n")

print("Count of values for each species:")
print(grouped.count()) 
```
